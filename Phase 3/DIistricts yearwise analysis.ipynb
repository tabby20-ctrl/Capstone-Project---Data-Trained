```python
#check for any null values
dff.isnull().sum()
```

```python
# select districts that appears 3 or mre than 3 years in the last decade
prop = dff['DISTRICT'].value_counts()
```

```python
prop1 = prop[prop.values >= 3]
```

```python
plt.figure(figsize=(10,14))
sns.barplot(x=prop1.values , y= prop1.index)
plt.title('appeared districts in maximum murder crimes over the last decade')
plt.show()
```
